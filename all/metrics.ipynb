```python
import requests
import json
import time
from bs4 import BeautifulSoup
import pandas as pd
import datetime

import plotly.express as px
import plotly.graph_objects as go
```

```python
dfs = []
for i in range(1,6):
  df = pd.read_csv("personal_info" + str(i) + ".csv")
  df["コンテスト"] = "皆解会" + str(i)
  df_a = df.copy()
  df_a["レート"] = df["Aレート"]
  df_a["タイプ"] = "アルゴリズム"
  df_h = df.copy()
  df_h["レート"] = df["Hレート"]
  df_h["タイプ"] = "ヒューリスティック"
  df = pd.concat([df_a,df_h])
  dfs.append(df)

df = pd.concat(dfs)
df
```

```python
fig = px.box(df, x="コンテスト", y="レート", color="タイプ", points='all', title="過去の皆解会参加者のレート")
fig.show()
```
